```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
im1 = cv2.imread('/content/star.jpg')
cv2_imshow(im1)
im2 = cv2.imread('/content/diamond.jpg')
cv2_imshow(im2)
im3 = cv2.add(im1,im2)
cv2_imshow(im3)
sub = cv2.subtract(im1,im2)
cv2_imshow(sub)
```

```python
input1 = cv2.imread('/content/black_and_white.png')
cv2_imshow(input1)
input2 = cv2.imread('/content/circle.png')
cv2_imshow(input2)

and_img = cv2.bitwise_and(input1,input2)
cv2_imshow(and_img)

or_img = cv2.bitwise_or(input1,input2)
cv2_imshow(or_img)

xor_img = cv2.bitwise_xor(input1,input2)
cv2_imshow(xor_img)

not_img = cv2.bitwise_not(input2)
cv2_imshow(not_img)

```

```python

image = cv2.imread('/content/testimage.jpg');
img = cv2.resize(image, (0, 0), fx = 0.4, fy = 0.4)

cv2_imshow(img);
```

```python
rotated_image = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated_image)

height, width = img.shape[:2]
center = (width/2,height/2)
rotate_matrix = cv2.getRotationMatrix2D(center=center, angle=45, scale=1)
rotated_image2 = cv2.warpAffine(src=img, M=rotate_matrix, dsize=(width, height))
cv2_imshow(rotated_image2)
```

```python
tx, ty = width / 4, height / 4
 
# create the translation matrix using tx and ty, it is a NumPy array 
translation_matrix = np.array([
    [1, 0, tx],
    [0, 1, ty]
], dtype=np.float32)

translated_image = cv2.warpAffine(src=img, M=translation_matrix, dsize=(width, height))
cv2_imshow(translated_image)
```

```python
img_v = cv2.flip(img,0)
cv2_imshow(img_v)

img_h = cv2.flip(img,1)
cv2_imshow(img_h)
```

```python
averaging = cv2.blur(img,(10,10)) 
cv2_imshow(averaging)

gausBlur = cv2.GaussianBlur(img, (5,5),0) 
cv2_imshow(gausBlur)
cv2.waitKey(0)
  
# Median blurring
medBlur = cv2.medianBlur(img,5)
cv2_imshow(medBlur)
cv2.waitKey(0)
  
# Bilateral Filtering
bilFilter = cv2.bilateralFilter(img,9,75,75)
cv2_imshow(bilFilter)


filter = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
sharpen_img=cv2.filter2D(gausBlur,-1,filter)
cv2_imshow(sharpen_img)
```

```python
edges = cv2.Canny(img,100,200)
cv2_imshow(edges)
```
